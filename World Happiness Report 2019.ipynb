# World Happiness Report 2019

It is a report which is published by Sustainable Developemnt Solutions Network.This report ranks 149 Countries by how happy their citizens perceive themselves to be.Basically ,respondants are asked to rate their own current lives on a scale of 0-10 during survey .Ranking is done on the basis of Six broad Parameters -GDP Per Capita,Social Support,Healthy life expectancy,Freedom to make life choices ,Generosity and Perceptions of Corruption .
Meanings of the term involved in it:
1. GDP Per Capita-It shows a country's GDP divided by its total Population.
2. Social Support-It means having family ,friends etc. to share their feelings .
3. Health-Healthy Life expectancy stands for average number of years that a person speds healthy.
4. Freedom-It is person's autonomy to perform action according to his own desire .
5. Generosity-It stands for how much a people is contributing to charity.
6. Corruption-It is the abuse of entrusted power .

```python
#Univariate Analysis
#Barplot between Overall Rank and Healthy life expectancy 
x = data['Overall rank'].head(5)
y = data['Healthy life expectancy'].tail(5)
 
# Figure Size
fig = plt.figure(figsize =(10, 7))
plt.title('Plot between Overall rank and Healthy life expectancy')
plt.xlabel('Overall rank',fontsize = 15)
plt.ylabel('Healthy life expectancy',fontsize = 15)


 
# Horizontal Bar Plot
plt.bar(x,y)
 
# Show Plot
plt.show()
```

```python
#Univariate analysis 
#Using Subplot
top_5 = data['Country or region'].head(5)
bottom_5 = data['Country or region'].tail(5)
```

```python
x1 = data['Country or region'].head(5)
y1 = data['GDP per capita'].head(5)

x2 = data['Country or region'].tail(5)
y2 = data['GDP per capita'].tail(5)

fig = plt.figure(figsize =(10, 7))
plot1 = plt.figure(1)
plt.xlabel("Name of Countries")
plt.ylabel("GDP Per Capita")
plt.title("Top five countries with their GDP Per capita")
plt.bar(x1,y1)

fig = plt.figure(figsize =(10, 7))
plot2 = plt.figure(2)
plt.xlabel("Name of Countries")
plt.ylabel("GDP Per capita")
plt.title("Bottom five countries with their GDP Per capita")
plt.bar(x2,y2)

plt.show()
```

```python
#Bivariate Analysis
#Scatterplot between Score and Corruption
plt.scatter(data['Perceptions of corruption'],data['Score'])
plt.xlabel("Happiness score")
plt.ylabel("Corruption")
plt.title("Happiness Vs Corruption")
plt.show()
```

```python
#Scatterplot between Overall Score and Generosity 
plt.scatter(data['Generosity'],data['Score'])
plt.xlabel("Happiness score")
plt.ylabel("Generosity")
plt.title("Happiness Vs Generosity")
plt.show()
```

```python
#Scatterplot between Overall Score  and Healthy life expectancy
plt.scatter(data['Healthy life expectancy'],data['Score'])
plt.xlabel("Happiness score")
plt.ylabel("Healthy life expectancy")
plt.title("Happiness Vs Healthy life expectancy")
plt.show()
```

```python
#Scatterplot between Overall score and Social support 
plt.scatter(data['Social support'],data['Score'])
plt.xlabel("Happiness score")
plt.ylabel("Social support")
plt.title("Happiness Vs Social support")
plt.show()
```
